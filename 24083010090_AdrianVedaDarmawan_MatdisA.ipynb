```python
import pandas as pd
import numpy as np
```

```python
kelasku = pd.read_csv('https://raw.githubusercontent.com/adriann03/dataset_mahasiswa_matdis_a/refs/heads/main/data_matdis_a.csv')
kelasku.head()
kelasku.info()
```

```python
columns_to_keep = kelasku.columns[0:6]
rows_to_keep = kelasku.index[0:40]
kelasku_filtered = kelasku[columns_to_keep]
kelasku_filtered = kelasku_filtered.loc[rows_to_keep]
display(kelasku_filtered)
```

```python
numerical_data = kelasku_filtered.drop(columns=['NPM', 'NAMA'])

cov_mat = np.cov(numerical_data, rowvar=False, ddof=1)
corr_mat = np.corrcoef(numerical_data, rowvar=False)

print("Covariance Matrix:")
display(pd.DataFrame(cov_mat, columns=numerical_data.columns, index=numerical_data.columns))

print("\nCorrelation Matrix:")
display(pd.DataFrame(corr_mat, columns=numerical_data.columns, index=numerical_data.columns))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim_matrix = cosine_similarity(numerical_data)

display(pd.DataFrame(cosine_sim_matrix, index=kelasku_filtered['NAMA'], columns=kelasku_filtered['NAMA']))
```
