Pernyataan yang diambil:
1. $1 + 2 + 2^2 + 2^3 + ... + 2^n = 2^{n+1} - 1$ untuk semua $n ∈ Z, n \geq 0$
2. $2^n > n^2$ untuk semua $n ∈ Z$


```python
def sisiKiri1(x):
  sum = 0
  for i in range(x+1):
    sum += 2**i
  return sum
def sisiKanan1(x):
  return 2**(x+1) - 1
def kebenaran1(kiri, kanan):
  return kiri == kanan
def ujiString1(x):
  return f"(n = {i}) {sisiKiri1(i)} ?= {sisiKanan1(i)} ({kebenaran1(sisiKiri1(i), sisiKanan1(i))})"


for i in range(10):
  print(ujiString1(i))
print()
for i in range(10, 100, 10):
  print(ujiString1(i))
print()
for i in range(100, 1000, 100):
  print(ujiString1(i))
```

Semua $n \geq 0$ berlaku pada pernyataan 1

```python
def sisiKiri2(x):
  return 2**x
def sisiKanan2(x):
  return x**2
def kebenaran2(kiri, kanan):
  return kiri > kanan
def ujiString2(x):
  return f"(n = {x}) {sisiKiri2(x)} ?> {sisiKanan2(x)} ({kebenaran2(sisiKiri2(x), sisiKanan2(x))})"


print("----- n >= 0 -----")
for i in range(10):
  print(ujiString2(i))
print()
for i in range(10, 100, 10):
  print(ujiString2(i))
print()
for i in range(100, 1000, 100):
  print(ujiString2(i))
print("\n"*4)
print("----- n < 0 -----")
for i in range(1, 10):
  print(ujiString2(-i))
print()
for i in range(10, 100, 10):
  print(ujiString2(-i))
print()
for i in range(100, 1000, 100):
  print(ujiString2(-i))
```

Tidak semua $n$ berlaku pada pernyataan 2. Yang tidak memenuhi pernyataan berada di $2 \leq n \leq 4$, dan $n < 0$
